# Plot Data dengan Matplotlib

# Import Modules

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Persamaan Garis

```python
x = np.arange(0,11,1)
y = 2 * x + 3 
```

```python
x
```

```python
y
```

```python
# plt.figure(1)
plt.plot(x,y)
plt.show()
```

# Lingkaran

```python
# plt.figure(1)
jari2 = 5
sudut = np.linspace(0,2 * np.pi,100)
x2 = jari2 * np.cos(sudut)
y2 = jari2 * np.sin(sudut)
```

```python
plt.plot(x2,y2)
plt.show()
```

# Multiple Subplots

```python
fig, (ax1,ax2) = plt.subplots(ncols=1,nrows=2)

ax1.plot(x,y,'ro-',label='Persamaan Garis')
ax1.set_ylabel('Persamaan Garis')

ax2.plot(x2,y2,'gs--',label='Lingkaran')
ax2.set_xlabel('X Axis')
ax2.set_ylabel('Lingkaran')

fig.suptitle('Revealing The Result')
fig.legend(loc='upper left')

plt.show()
```
