# Animasi Pada Matplotlib

```python
fig = plt.figure()
plt.xlim((0,4))
plt.ylim((-2,2))

line = plt.plot([])[0]

def animate(i):
    x = np.linspace(0,4,1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x,y)
    return line,

anim = FuncAnimation(fig,animate,frames=300,interval=30)

video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)

plt.close()
```
