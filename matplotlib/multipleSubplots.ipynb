# Simple Line Plots Using OO Style

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,2,0.01)
```

```python
print(x)
```

```python
s = np.sin(2 * np.pi * x)
```

```python
print(s)
```

```python
fig, ax = plt.subplots()

ax.plot(x,s)

ax.set(xlabel='Nilai X',ylabel='Nilai Sin (X)',title='Visualisasi Nilai Sin')

ax.grid()
plt.show()
```

# Simple Line Plot Menggunakan Pyplot Style

```python
plt.plot(x,s)

plt.title('Visualisasi Nilai Sin')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Sin(X)')

plt.grid()
plt.show()
```

# Multiple Subplots Menggunakan OO Style

```python
x1 = np.linspace(0,5,100)
x2 = np.linspace(0,2,100)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
x1
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=2,ncols=1)

ax1.plot(x1,y1,marker='s',linestyle='--',c='r')
ax2.plot(x2,y2,marker='o',linestyle='-',c='b')

fig.suptitle('Multiple Subplots')
ax1.set_ylabel('Nilai $cos(x_1)$')
ax2.set_ylabel('Nilai $cos(x_2)$')
ax2.set_xlabel('Nilai X')

plt.show()
```

# Multiple Subplots Menggunakan Pyplot Style

```python
plt.subplot(2,1,1)
plt.plot(x1,y1)

plt.title('Multiple Subplots Menggunakan Pyplot Style')
plt.ylabel('Nilai $cos(x_1)$')

plt.subplot(2,1,2)
plt.plot(x2,y2)

plt.ylabel('Nilai $cos(x_2)$')
plt.xlabel('Nilai X')

plt.show()
```
