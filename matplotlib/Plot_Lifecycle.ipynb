# import modules

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = {
    'A':109438,
    'B':103569,
    'C':112214,
    'D':112591,
    'E':100934,
    'F':103660,
    'G':137351,
    'H':123381,
    'I':135841,
    'J':104437
}
df = pd.Series(data)
df
```

```python
items = tuple(data.keys())
items
```

```python
count = tuple(data.values())
count
```

# Simple OO Style Bar Horizontal Plot

```python
fig, ax = plt.subplots()

ax.barh(items,count)

plt.show()
```

# Pengaturan Style

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8')

fig, ax = plt.subplots()

ax.barh(items,count)

plt.show()
```

# Pengaturan Tick Labels

```python
fig, ax = plt.subplots()

ax.barh(items,count)

labels = ax.get_xticklabels()
plt.setp(labels,rotation=45,horizontalalignment='right')

plt.show()
```

# Pengaturan Format Pada Ticker

```python
from matplotlib.ticker import FuncFormatter
```

```python
def ribuan(x,pos):
    return f"{int(x/1000)}k"
```

```python
fig, ax = plt.subplots()

ax.barh(items,count)
formatter = FuncFormatter(ribuan)
ax.xaxis.set_major_formatter(formatter)

ax.set(xlabel='jumlah',ylabel='count',title='contoh plot')

plt.show()
```

# Pengaturan label Pada Sumbu (axis) Dan Judul (title)

```python
fig, ax = plt.subplots()

ax.barh(items,count)
ax.set(xlabel='jumlah',ylabel='count',title='contoh plot')

plt.show()
```

# Penambahan Garis (vertical/horizontal line) pada plot

```python
fig, ax = plt.subplots()

ax.barh(items,count)

ax.axvline(np.mean(count),ls='--',c='r')

ax.set(title='contoh plot',xlabel='jumlah',ylabel='count')

plt.show()
```
