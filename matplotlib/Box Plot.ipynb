# Box Plot

# Import Modules

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Sample Dataset

```python
np.random.seed(2)
data = np.random.normal(loc=100,scale=10,size=200)
data
```

# Simple Box Plot OO Style

```python
fig, ax = plt.subplots()

ax.boxplot(data,labels=['Data'])

ax.set(title='Simple Box Plot Using OO Style',xlabel='X Axis',ylabel='Y Axis')

ax.grid()
plt.show()
```

# Simple Box Plot Using Pyplot Style

```python
plt.boxplot(data,labels=['Data'])

plt.title('Simple Box Plot Using Pyplot Style')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.grid()
plt.show()
```

# Notched Box Plot

```python
plt.boxplot(data,labels=['Data'],notch=True)

plt.title('Simple Box Plot Using Notch Parameter')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.grid()
plt.show()
```

# Pengaturan Marker Pada Outlier

```python
plt.boxplot(data,labels=['Data'],showfliers=True,flierprops={'markerfacecolor':'r','marker':'*'})

plt.title('Customizing The Box Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.grid()
plt.show()
```
