# Import Modules

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Mengeluarkan 5 bilangan float acak antara 0 dan 1
data1 = np.random.uniform(size=5)
print(data1)
```

```python
# Mengeluarkan 5 bilangan float acak dari distribusi normal dengan mean 0 dan std 1
data2 = np.random.normal(loc=10, scale=2, size=100)
print(data2)
```

```python
x = np.random.randint(low=1, high=13, size=100)
x
```

```python
y = np.random.normal(loc=10,scale=2,size=100)
y
```

# Simple Scatter Plot

```python
plt.scatter(x,y,label='data 1',c='r')

plt.title('Contoh Scatter Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.legend()
plt.show()
```

# Multiple Scatter Plot

```python
x1 = np.random.randint(low=1,high=15,size=15)
y1 = np.random.normal(loc=10,scale=2,size=15)
```

```python
x2 = np.random.normal(loc=15,scale=1,size=15)
y2 = np.random.randint(low=1,high=15,size=15)
```

```python
plt.scatter(x1,y1,label='data 1',c='r')
plt.scatter(x2,y2,label='data 2',c='b')

plt.title('Multiple Scatter Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.legend()
plt.show()
```

# Pengaturan Marker

```python
x1 = np.random.randint(low=1,high=20,size=20)
y1 = np.random.normal(loc=10,scale=3,size=20)
```

```python
x2 = np.random.normal(loc=15,scale=2,size=20)
y2 = np.random.randint(low=1,high=20,size=20)
```

```python
plt.scatter(x1,y1,c='r',marker='s',linewidths=2,edgecolor='black',s=200,label='data 1')
plt.scatter(x2,y2,c='b',marker='^',linewidths=3,edgecolor='black',s=400,label='data 2')

plt.title('Cutomizing Scatter Plot')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')

plt.legend()
plt.show()
```

# Scatter plot untuk multi dimensi

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
features = iris['data'].T
```

```python
iris.keys()
```

```python
iris['data']
```

```python
iris['feature_names']
```

```python
features = iris['data'].T
features
```

```python
iris['target']
```

```python
iris['target_names']
```

```python
plt.scatter(features[0],features[1],alpha=0.2,s=100*features[3],c=iris['target'],cmap='viridis')

plt.title('iris dataset')
plt.xlabel(iris['feature_names'][0])
plt.ylabel(iris['feature_names'][1])

plt.colorbar()
plt.show()
```
