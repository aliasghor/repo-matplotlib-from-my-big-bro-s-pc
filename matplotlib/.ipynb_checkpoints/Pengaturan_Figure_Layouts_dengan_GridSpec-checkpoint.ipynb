# Pengaturan Figure Layouts dengan GridSpec

# Import Modules

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
import pandas as pd
```

# Simple 2x2 layout dengan subplots

```python
nama_siswa = ['Gerry','Gendut','Pace','Mogi']
uas = [90,45,80,75]

fig, ax = plt.subplots(nrows=2,ncols=2,constrained_layout=True)

ax[0][0].bar(nama_siswa,uas)
ax[0][1].pie(uas,labels=nama_siswa,autopct='%1.1f%%',wedgeprops={'edgecolor':'black'},explode=[0.1,0,0,0])
ax[1][0].pie(uas,labels=nama_siswa,autopct='%1.1f%%',wedgeprops={'edgecolor':'black'})
ax[1][1].barh(nama_siswa,uas)

plt.show()
```

# Simple 2x2 Layout Dengan Gridspec

```python
fig = plt.figure(constrained_layout=True)
gs = gridspec.GridSpec(ncols=2,nrows=2,figure=fig)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

ax1.bar(nama_siswa,uas)
ax2.pie(uas,labels=nama_siswa,autopct='%1.1f%%',wedgeprops={'edgecolor':'black'})
ax3.pie(uas,labels=nama_siswa,autopct='%1.1f%%',wedgeprops={'edgecolor':'black'},explode=[0.1,0,0,0])
ax4.barh(nama_siswa,uas)

plt.show()
```

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(ncols=2,nrows=2)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

colors = ['purple','red','yellow','gold']

ax1.bar(nama_siswa,uas)
ax2.pie(uas,labels=nama_siswa,autopct='51.1f%%',colors=colors,wedgeprops={'edgecolor':'black'},explode=[0.1,0,0,0])
ax3.pie(uas,labels=nama_siswa,autopct='%1.1f%%',colors=colors,wedgeprops={'edgecolor':'black'})
ax4.barh(nama_siswa,uas)

plt.show()
```

# Pengaturan Rows Span Dan Cols Span Pada Grid Spec

```python
fig = plt.figure(constrained_layout=True)
gs = gridspec.GridSpec(nrows=2,ncols=2,figure=fig)

ax1 = fig.add_subplot(gs[0,:])
# ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

ax1.bar(nama_siswa,uas)
ax2.pie(uas,labels=nama_siswa,autopct='51.1f%%',colors=colors,wedgeprops={'edgecolor':'black'},explode=[0.1,0,0,0])
ax3.pie(uas,labels=nama_siswa,autopct='%1.1f%%',colors=colors,wedgeprops={'edgecolor':'black'})
ax4.barh(nama_siswa,uas)

plt.show()
```
