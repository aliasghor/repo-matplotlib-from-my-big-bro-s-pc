# Bar Plot

# Import Modules

```python
%matplotlib inline
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
huruf = list('ABCDE')
huruf
```

```python
angka = [10,30,20,50,70]
angka
```

# Simple Bar Plot

```python
plt.bar(huruf,angka)

plt.title('Simple bar Plot With Pyplot Style')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```

# Pengaturan Grid Dan Warna Pada Bar Plot

```python
plt.bar(huruf,angka,color='g',alpha=0.50)

plt.title('Customizing Bar Plot')
plt.xlabel('Kategori')
plt.ylabel('Jumlah')

plt.grid(linestyle='--',c='r',linewidth=2,axis='y')
plt.show()
```

# Horizontal Bar Plot

```python
plt.barh(huruf,angka)

plt.title('Horizontal Bar Horizontal plot')
plt.xlabel('Kategori')
plt.ylabel('Jumlah')

plt.grid(linestyle='--',linewidth=2,color='r',axis='x')
plt.show()
```

# Reversing Horizontal Bar Plot

```python
plt.barh(huruf[::-1],angka[::-1])

plt.title('Horizontal Bar Horizontal plot')
plt.xlabel('Kategori')
plt.ylabel('Jumlah')

plt.grid(linestyle='--',linewidth=2,color='r',axis='x',alpha=0.75)
plt.show()
```

# Stacked Bar Plot

```python
data1 = [25,85,75,40,60]
data2 = [40,35,20,55,10]
kategori =['A','B','C','D','E']
```

```python
plt.bar(kategori,data1,label='data 1')
plt.bar(kategori,data2,label='data 2',bottom=data1)

plt.title('Contoh Stacked Bar Plot')
plt.xlabel('Kategori')
plt.ylabel('Jumlah')

plt.grid(linestyle='--',linewidth=2,color='r',axis='y',alpha=0.75)
plt.legend()
plt.show()
```

```python
plt.bar(kategori,data1,label='data 1',bottom=data2)
plt.bar(kategori,data2,label='data 2')

plt.title('Contoh Stacked Bar Plot')
plt.xlabel('Kategori')
plt.ylabel('Jumlah')

plt.grid(linestyle='--',linewidth=2,color='r',axis='y',alpha=0.75)
plt.legend()
plt.show()
```

# Groupped Bar Plot

```python
data1 = [25,85,75,40,60]
data2 = [40,35,20,55,10]
kategori =['A','B','C','D','E']
x = np.arange(len(kategori))
x
```

```python
width = 0.35

plt.bar(x-width/2,data1,width,label='data 1')
plt.bar(x+width/2,data2,width,label='data 2')

plt.title('Groupped Bar Plot')
plt.xlabel('Kategori')
plt.ylabel('Jumlah')

plt.grid(linestyle='--',c='r',axis='y',linewidth=2)
plt.legend()
plt.show()
```
