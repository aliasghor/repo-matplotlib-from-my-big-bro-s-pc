```python
import matplotlib.pyplot as plt
import numpy as np
```

# membuat 3D axes

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
plt.show()
```

# 3D line plot

```python
z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(x,y,z,c='r')
ax.set(xlabel='Sumbu X',ylabel='Sumbu Y',zlabel='Sumbu Z',title='Contoh 3D line plot')
plt.show()
```

# 3D Scatter plot

```python
z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter(x,y,z,c='y')
ax.set(xlabel='Sumbu X',ylabel='Sumbu Y',zlabel='Sumbu Z',title='3D Scatter Plot')
plt.show()
```
