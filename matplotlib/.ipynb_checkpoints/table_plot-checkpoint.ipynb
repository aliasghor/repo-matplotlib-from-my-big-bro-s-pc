# Ngebuat Table Plot

# Magic Command

```python
%matplotlib inline
```

# Import Module

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Sample Data

```python
data_uas = [
    ["Gerry",70],
    ["Sampi",83],
    ["Mogi",62],
    ["Pace",74],
    ["Gendut",71]
]
```

```python
data_uas
```

# Table Plot Dengan OO Style

```python
fig, ax = plt.subplots()

table = plt.table(cellText=data_uas,loc="center")

table.set_fontsize(18)
table.scale(1,4) # Ukuran Kolom Dan Baris

ax.axis(False)
plt.show()
```

# Table Plot Menggunakan Pyplot Style

```python
table = plt.table(cellText=data_uas,loc="center")

table.set_fontsize(18)
table.scale(2,5)

ax = plt.gca()
ax.axis(False)

plt.show()
```

```python
fig, ax = plt.subplots()

table = plt.table(cellText=data_uas,loc="center")

table.set_fontsize(18)
table.scale(1.5,4.5)

ax.axis(False)
plt.show()
```

```python
table = plt.table(cellText=data_uas,loc="center")

table.set_fontsize(14)
table.scale(0.5,3.5)

ax = plt.gca()
ax.axis(False)

plt.show()
```
