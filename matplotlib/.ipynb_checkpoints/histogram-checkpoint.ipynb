# Histogram

# Import Modules

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Sample Data set

```python
mu, sigma = 100, 15 # Nilai Mean Dan Nilai Standart Deviation
x = mu + sigma * np.random.randn(10000) # Normal Distribution
x
```

```python
x.shape
```

# Histogram Dengan Pyplot Style

```python
plt.hist(x,bins=50,facecolor='g',alpha=0.75)

plt.title('Histogram With Pyplot Style')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.text(46,525,r'$\mu=100,\ \sigma=15$')
plt.grid()
plt.show()
```

# Histogram Dengan OO Style

```python
fig, ax = plt.subplots()

ax.hist(x,bins=50,facecolor='y',alpha=0.50)

ax.set(title='Histogram With OO Style',xlabel='X Axis',ylabel='Y Axis')

ax.text(46,525,'$\mu=100,\ \sigma=15$')

ax.grid()
plt.show()
```
