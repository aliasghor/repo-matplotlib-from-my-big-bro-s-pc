```python
import matplotlib.pyplot as plt
import numpy as np
```

# sample dataset

```python
n = np.linspace(-3.0,3.0,100)
x, y = np.meshgrid(n,n)
z = np.sqrt(x**2 + y**2)
print(x.shape)
print(y.shape)
print(z.shape)
```

# contour plot

```python
plt.contour(x,y,z)
plt.colorbar()

plt.title('contoh contour plot')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')

plt.show()
```

# filled contour plot

```python
plt.contourf(x,y,z)
plt.colorbar()

plt.title('filled countour plot')
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')

plt.show()
```
