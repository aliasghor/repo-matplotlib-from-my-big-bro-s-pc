```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
x = np.linspace(1,10,30)
x
```

```python
fig, ax1 = plt.subplots()

ax1.plot(x,np.exp(x),'ro-',label='exponent')
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('Exponent')

ax2 = ax1.twinx()
ax2.plot(x,np.log(x),'c^--',label='Log')
ax2.set_ylabel('Log')

fig.suptitle('Twin Axes')
fig.legend(loc='upper left')

plt.show()
```
