```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = np.linspace(-3.0,3.0,100)
x, y = np.meshgrid(n,n)
z = np.sqrt(x**2 + y**2)
```

```python
plt.contour(x,y,z)

plt.title('Contour Chart')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.colorbar()
plt.show()
```

```python
plt.contourf(x,y,z)

plt.title('Filled Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.colorbar()
plt.show()
```
