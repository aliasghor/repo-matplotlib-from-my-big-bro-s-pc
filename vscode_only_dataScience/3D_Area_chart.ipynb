```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = np.linspace(-3.0,3.0,100)
x, y = np.meshgrid(n,n)
z = np.sqrt(x**2 + y**2)
```

```python
print(x.shape)
print(y.shape)
print(z.shape)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(x,y,z,color='purple')
ax.set(title='3D Wire Frame',xlabel='Sumbu X',ylabel='Sumbu Y',zlabel='Sumbu Z')

plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x,y,z,cmap='plasma')
ax.set(title='3D Surface Plot',xlabel='Sumbu X',ylabel='Sumbu Y',zlabel='Sumbu Z')

plt.show()
```
