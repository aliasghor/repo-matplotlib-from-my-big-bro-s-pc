```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-bright')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(x,y,z)
ax.set(title='3D Line Plot',xlabel='Sumbu X',ylabel='Sumbu Y',zlabel='Sumbu Z')

plt.show()
```

```python
z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter(x,y,z)
ax.set(title='3D Scatter Plot',xlabel='Sumbu X',ylabel='Sumbu Y',zlabel='Sumbu Z')

plt.show()
```
