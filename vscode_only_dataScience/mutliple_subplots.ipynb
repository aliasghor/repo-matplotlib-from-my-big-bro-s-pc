```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
x = np.arange(0,2,0.01)
x
```

```python
y = np.sin(2 * np.pi * x)
y
```

```python
fig, ax = plt.subplots()

ax.plot(x,y,'ro--')
ax.set(title='Visualisasi Nilai Sin',xlabel='Sumbu X',ylabel='Sumbu Y')
ax.grid()

plt.show()
```

```python
x1 = np.linspace(0,5,100)
x2 = np.linspace(0,2,100)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=1,nrows=2)

ax1.plot(x1,y1,'c^--',label='nilai $(x_1)$')
ax1.set_ylabel('Nilai $(x_1)$')

ax2.plot(x2,y2,'ro-',label='nilai $(x_2)$')
ax2.set_xlabel('Sumbu X')
ax2.set_ylabel('Nilai $(x_2)$')

fig.suptitle('Mutliple Subplots')
fig.legend(loc='upper left')

plt.show()
```
