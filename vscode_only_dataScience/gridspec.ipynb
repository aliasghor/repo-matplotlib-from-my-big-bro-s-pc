```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-bright')
```

```python
devs = ["Python","C++","JavaScript","Java"]
salary = [5500,4500,5000,4000]
```

# 2x2 Layout With Gridspec

```python
fig = plt.figure(constrained_layout=True)
gs = gridspec.GridSpec(ncols=2,nrows=2,figure=fig)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

ax1.bar(devs,salary)
ax2.pie(salary,labels=devs,autopct="%1.1f%%",wedgeprops={'edgecolor':'black'},explode=[0.1,0,0,0])
ax3.pie(salary,labels=devs,autopct="%1.1f%%",wedgeprops={'edgecolor':'black'},)
ax4.barh(devs,salary)

plt.show()
```

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(ncols=2,nrows=2)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

ax1.bar(devs,salary)
ax2.pie(salary,labels=devs,autopct="%1.1f%%",wedgeprops={'edgecolor':'black'},explode=[0.1,0,0,0])
ax3.pie(salary,labels=devs,autopct="%1.1f%%",wedgeprops={'edgecolor':'black'})
ax4.barh(devs,salary)

plt.show()
```

# 2x2 Layout With Subplots

```python
fig, ax = plt.subplots(ncols=2,nrows=2,constrained_layout=True)

ax[0][0].bar(devs,salary)
ax[0][1].pie(salary,labels=devs,autopct="%1.1f%%",wedgeprops={'edgecolor':'black'},explode=[0.1,0,0,0])
ax[1][0].pie(salary,labels=devs,autopct="%1.1f%%",wedgeprops={'edgecolor':'black'})
ax[1][1].bar(devs,salary)

plt.show()
```
