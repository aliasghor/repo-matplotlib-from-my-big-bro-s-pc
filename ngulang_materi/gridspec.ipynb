```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
```

```python
plt.style.use('seaborn-v0_8-bright')
```

# 2x2 layout with gridspec

# Dataset

```python
programming_languages = ['Python','C++','JavaScript','Java','C']
salaries = [6500,5500,6000,5000,4500]
```

```python
fig = plt.figure(constrained_layout=True)
gs = gridspec.GridSpec(ncols=2,nrows=2,figure=fig)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

labels = ax1.get_xticklabels()

ax1.bar(programming_languages,salaries)
plt.setp(labels,rotation=35,horizontalalignment='right')
ax2.pie(salaries,labels=programming_languages,wedgeprops={"edgecolor":"black"},autopct='%1.1f%%',explode=[0.1,0,0,0,0])
ax3.pie(salaries,labels=programming_languages,wedgeprops={"edgecolor":"black"},autopct='%1.1f%%')
ax4.barh(programming_languages,salaries)

plt.show()
```
