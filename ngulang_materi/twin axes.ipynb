# Twin Axes

# Import Modules

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Sample Dataset

```python
x = np.linspace(1,10,30)
x
```

# Twin Axes

```python
fig, ax1 = plt.subplots()

ax1.plot(x,np.exp(x),'ro-',label='Exponent')
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('Exponent')

ax2 = ax1.twinx()
ax2.plot(x,np.log(x),'y^--',label='Log')
ax2.set_ylabel('Log')

fig.suptitle('Twin Axes')
fig.legend(loc='upper left')

plt.show()
```
