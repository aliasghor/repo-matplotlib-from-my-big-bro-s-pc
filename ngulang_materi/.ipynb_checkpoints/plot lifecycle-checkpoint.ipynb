# Plot Lifecycle

# Import Module

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    'A':109438,
    'B':103569,
    'C':112214,
    'D':112591,
    'E':100934,
    'F':103660,
    'G':137351,
    'H':123381,
    'I':135841,
    'J':104437
}
```

```python
data_keys = tuple(data.keys())
data_keys
```

```python
data_values = tuple(data.values())
data_values
```

```python
fig, ax = plt.subplots()

ax.barh(data_keys,data_values)
ax.set(xlabel='Sumbu X',ylabel='Sumbu Y',title='OO Stlye Horizontal Bar Plot')

plt.show()
```

```python
fig, ax = plt.subplots()

labels = ax.get_xticklabels()

ax.barh(data_keys,data_values)
ax.set(xlabel='Sumbu X',ylabel='Sumbu Y',title='OO Stlye Horizontal Bar Plot')
plt.setp(labels,rotation=45,horizontalalignment='right')

plt.show()
```

```python
fig, ax = plt.subplots()

labels = ax.get_xticklabels()

ax.barh(data_keys,data_values)
ax.set(xlabel='Sumbu X',ylabel='Sumbu Y',title='OO Stlye Horizontal Bar Plot')
ax.axvline(np.mean(data_values),ls='--',c='r')
plt.setp(labels,rotation=45,horizontalalignment='right')

plt.show()
```
