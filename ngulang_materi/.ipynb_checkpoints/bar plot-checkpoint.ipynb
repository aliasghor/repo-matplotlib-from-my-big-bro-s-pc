# Bar Plot

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
salaries = [5000,4500,3000,3500]
programming_languages = ['Python','JavaScript','C++','Java']
```

```python
plt.bar(programming_languages,salaries)

plt.title('Simple Bar Plot With Pyplot Style')
plt.xlabel('Developer')
plt.ylabel('Salaries')

plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(programming_languages,salaries)
ax.set(title='Simple Bar Plot With OO Style',xlabel='Developer',ylabel='Salaries')

ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.barh(programming_languages,salaries)
ax.set(title='Simple Bar Plot With OO Style',xlabel='Developer',ylabel='Salaries')

ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.barh(programming_languages,salaries,color='red')
ax.set(title='Simple Bar Plot With OO Style',xlabel='Developer',ylabel='Salaries')

ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(programming_languages,salaries)
ax.set(title='Simple Bar Plot With OO Style',xlabel='Developer',ylabel='Salaries')

ax.grid(linestyle='--',linewidth=2,axis='y',color='red',alpha=0.75)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.barh(programming_languages,salaries)
ax.set(title='Simple Bar Plot With OO Style',xlabel='Developer',ylabel='Salaries')

ax.grid(linestyle='--',linewidth=2,axis='x',color='red',alpha=0.75)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(programming_languages,salaries)
ax.set(title='Customizing Bar Plot',xlabel='Devs',ylabel='Salaries')
ax.grid(ls='--',axis='y',c='r',alpha=0.75,linewidth=2)

labels = ax.get_xticklabels()

plt.setp(labels,horizontalalignment='right',rotation=45)
plt.show()
```
