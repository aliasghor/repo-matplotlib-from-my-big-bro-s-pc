# Import modules

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [2,4,6,8,10,11,11.5,11.7]
y = [1,1.5,2,2.5,3,3.5,4,4.5]
```

```python
plt.scatter(x,y,label='Data 1',c='r')

plt.title('Simple Scatter Plot With Pyplot Style')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```
