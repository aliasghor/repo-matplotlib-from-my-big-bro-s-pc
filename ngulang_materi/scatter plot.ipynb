```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [2,4,6,8,10,11,11.5,11.7]
y = [1,1.5,2,2.5,3,3.5,4,4.5]
```

```python
plt.scatter(x,y,label='Data 1',c='r')

plt.title('Simple Scatter Plot With Pyplot Style')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()
plt.show()
```

```python
x1 = [2,4,6,8,10,11,11.5,11.7]
y1 = [1,1.5,2,2.5,3,3.5,4,4.5]

x2 = [8,8.5,9,9.5,10,10.5,11]
y2 = [3,3.5,3.7,4,4.5,5,5.2]
```

```python
plt.scatter(x1,y1,label='Data 1')
plt.scatter(x2,y2,label='Data 2')

plt.title('Multiple Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()
plt.show()
```

```python
plt.scatter(x1,y1,c='c',marker='^',edgecolor='red',s=200,label='Data 1')
plt.scatter(x2,y2,c='c',marker='*',edgecolor='blue',s=400,label='Data 2')

plt.title('Multiple Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()
plt.show()
```
