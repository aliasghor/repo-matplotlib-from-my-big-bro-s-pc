# Contour Chart

# Import Modules

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = np.linspace(-3.0,3.0,100)
x, y = np.meshgrid(n,n)
z = np.sqrt(x**2 + y**2)
print(x.shape)
print(y.shape)
print(z.shape)
```

```python
plt.contour(x,y,z)
plt.colorbar()

plt.title('Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

# Filled Countour Plot

```python
n = np.linspace(-3.0,3.0,100)
x, y = np.meshgrid(n,n)
z = np.sqrt(x**2 + y**2)
```

```python
plt.contourf(x,y,z)
plt.colorbar()

plt.title('Filled Countour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
